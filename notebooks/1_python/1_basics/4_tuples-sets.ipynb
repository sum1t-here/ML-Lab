#### Set operations
union
intersection
difference
symmetric difference: returns elements that are present in either of sets but not both

```python
s1 = {"hiking", "cooking", "dancing", "coding"}
s2 = {"coding", "photography", "travel"}
```

```python
# intersection -> &
s1 & s2
```

```python
# union -> |
s1 | s2
```

```python
# difference -> -
s1 - s2
```

```python
# symmetric difference
s1 ^ s2
```
