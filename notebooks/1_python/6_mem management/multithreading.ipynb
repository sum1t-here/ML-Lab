```python
# Multithreading is a programming technique in which multiple threads execute concurrently within a single process to improve performance and resource utilization.

# A thread is the smallest unit of execution in a program. Multiple threads can exist within the same process and share its resources.

# A process is an independent program in execution with its own memory space. A process may contain multiple threads.
```

```python
import time
start = time.perf_counter()

def test_func():
    print("do something")
    print("sleep for 1 sec")
    time.sleep(1)
    print("done with sleeping")

# sequnetial
test_func()
test_func()
end = time.perf_counter()

print(f"The program is finished in {round(end-start, 2)} seconds.")
```

```python
import time
import threading
start = time.perf_counter()

def test_func():
    print("do something")
    print("sleep for 1 sec")
    time.sleep(1)
    print("done with sleeping \n")
    

# threads = []
# for i in range(10):
#     t = threading.Thread(target=test_func)
#     t.start()
#     threads.append(t)

# for thread in threads:
#     thread.join()

# concurrent
t1 = threading.Thread(target=test_func)
t2 = threading.Thread(target=test_func)

t1.start()  # to start the thread
t2.start()

t1.join()   # in order to first execute the threads and then the execution of the main program
t2.join()

end = time.perf_counter()

print(f"The program is finished in {round(end-start, 2)} seconds.")
```
