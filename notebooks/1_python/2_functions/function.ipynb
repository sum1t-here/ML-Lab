```python
# a function that take list as input and return all numeric value
def only_numeric(a):
    n = []
    for i in a:
        if type(i) == int or type(i) == float:
            n.append(i)
    return n
```

```python
only_numeric([1, 2, 3, 4, 5, 6])
```

```python
only_numeric([1, 2, 3, 4.5, 5, [1, 2, 3, 4,], "Machine learning"])
```

```python
# modified
def only_numeric(a):
    n = []
    for i in a:
        if type(i) == list:
            for j in i:
                if type(j) == int or type(j) == float:
                    n.append(j)
        else:
            if type(i) == int or type(i) == float:
                n.append(i)
    return n
```

```python
only_numeric([1, 2, 3, 4.5, 5, [1, 2, 3, 4,], "Machine learning"])
```

```python
# it should accept any number of arguments
# treated as key vlaue pair

def kwarg_func(**kwargs):
    return kwargs
```

```python
kwarg_func(a = 2, b = 3, c = [1,2,3])
```

```python
# returns a value which is a list

def test(**kwargs):
    for i in kwargs.keys():
        if type(kwargs[i]) == list:
            return i, kwargs[i]
```

```python
test(a = 2, b = 3, c = [1,2,3])
```

```python
def test(*arg, **kwargs):
    """ This function returns arg and kwargs
    Args:
    Kwargs
    """
    return arg, kwargs
```

```python
test(1,2,4, a=100, c=200)
```

```python
# factorial of a num
def fact_num(n):
    if n==0:
        return 1
    else:
        return n*fact_num(n-1)
```

```python
fact_num(5)
```

```python
fact_num = lambda n: 1 if n == 0 else n*fact_num(n-1)
```

```python
fact_num(5)
```

```python
# reduce -> folding/reduction

from functools import reduce
```

```python
l = [1, 2, 3, 4, 5, 6]
reduce(lambda x, y: x+y, l)
```

```python
def add(x,y):
    return x+y
reduce(add, l)
```

```python
reduce(lambda x, y: x*y, l)
```

```python
words = ["Data", "Science", "Course"]
reduce(lambda x, y : x + " " + y, words)
```

```python
def factorial(n):
    return reduce(lambda x, y: x*y, range(1, n+1))
```

```python
factorial(5)
```
