```python
# generator function -> type of function which doesnot return a single value, instead returns an iterator object

def sq_num(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result
sq_num(5)
```

```python
def sq_num_gen(n):
    for i in range(n):
        yield i ** 2
```

```python
gen = sq_num_gen(5)
```

```python
gen
```

```python
next(gen)
```

```python
next(gen)
```

```python
next(gen)
```

```python
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
list(fib(5))
```
