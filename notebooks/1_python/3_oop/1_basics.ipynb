`self` refers to the specific object (instance) that the method is being called on.

When you write:

```jsx
class Car:
    def accelerate(self):
        print("car is accelerating")
```


and then:

my_car = Car()
my_car.accelerate()


Python automatically converts this call to:

Car.accelerate(my_car)


So `self` inside the method will represent my_car.

Why do we need self?

- Stores attributes unique to each object
- Calls methods on the same object
- Python passes it automatically â€” we just have to include it in method definitions

````python
class Car:
    def accelerate(self): # methods / functions
        print("car is accelerating")
````

````python
a = Car()
````

````python
a.accelerate()
````
