```python
# Inheritance

# syntax
# class BaseClass:
#     body of base class
# class DerivedClass(BaseClass):
#     body of derived class
```

```python
# single inheritance -> when a class has only one parent class

class Father:
    def father_property(self):
        print("Father's property")
```

```python
class Son(Father):
    def job(self):
        print("Son have a job")
```

```python
child_obj = Son()
```

```python
child_obj.job()
```

```python
child_obj.father_property()
```

```python
class Fruit:
    def fruit_into(self):
        print("Inside Parent Class")
class Apple(Fruit):
    def fruit_info(self):
        print("Inside Child Class")
```

```python
apple = Apple()
apple.fruit_info() # method overriding
```

```python
# multi level inheritance
class GrandFather:
    def prop_grand_father(self):
        print("Highway wali zameen")
```

```python
class Father(GrandFather):
    def prop_father(self):
        print("I am your father")
```

```python
class Son(Father):
    def prop_son(self):
        print("Sbki property meri")
```

```python
son = Son()
```

```python
son.prop_father()
```

```python
son.prop_grand_father()
```

```python
son.prop_son() 
```

```python
# Multiple Inheritance -> one child may inherit from multiple parent class

class ParentClass1:
    def method1(self):
        print("I am from class 1")
class ParentClass2:
    def method2(self):
        print("I am from class 2")
```

```python
class ChildClass(ParentClass1, ParentClass2):
    def child_method(self):
        print("Child method")
```

```python
child = ChildClass()
child.method1()
```

```python
child.method2()
```

```python
child.child_method()
```

```python
class A:
    def method(self):
        print("Method of class A")
class B(A):
    def method(self):
        print("Method of class B")
class C(A):
    def method(self):
        print("Method of class C")
# class D(B, C):
#     pass
class D(C, B):
    pass
```

```python
d = D() # diamond problem
```

```python
d.method()
```

```python
d.method()
```

```python
# Heirarchial Inheritance

class Vehicle:
    def info(self):
        print("This is a vehicle")
class Car(Vehicle):
    def car_info(self, name):
        print("This car info", name)
class Truck(Vehicle):
    def truck_info(self, name):
        print("This truck info", name)
```

```python
obj1 = Car()
```

```python
obj1.info()
```

```python
obj1.car_info("BMW")
```

```python
obj2 = Truck()
```

```python
obj2.info()
```

```python
obj2.truck_info("TATA")
```

```python
# Hybrid inheritance

class Vehicle:
    def vehicle_info(self):
        print("Inside vehicle class")
class Car(Vehicle):
    def car_info(self):
        print("Inside car class")
class Truck(Vehicle):
    def truck_info(self):
        print("Inside truck class")

class SportsCar(Car, Vehicle):
    def sports_car(self):
        print("Inside sports car class")
```

```python
s_car = SportsCar()
```

```python
s_car.car_info()
```

```python
s_car.vehicle_info()
```

```python
s_car.sports_car()
```

```python
# Abstraction -> concept of hiding complex applications

import abc
```

```python
class Pwskills:
    
    @abc.abstractmethod
    def student_details(self):
        pass
    
    @abc.abstractmethod
    def student_assignment(self):
        pass
    
    @abc.abstractmethod
    def student_marks(self):
        pass
```

```python
class DataScience(Pwskills):
    def student_details(self):
        return "data science course student details"
    def student_marks(self):
        return "data science course student marks"
```

```python
class WebDev(Pwskills):
    def student_details(self):
        return "web dev course student details"
    def student_marks(self):
        return "web dev course student marks"
```

```python
ds = DataScience()
```

```python
ds.student_details()
```

```python
ds.student_assignment()
```
