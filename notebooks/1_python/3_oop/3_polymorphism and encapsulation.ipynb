```python
# Polymorphism -> method overriding, method overloading

# method overloading -> the last method overwrites the previous ones
class Student:
    def student(self):
        print("Welcome to pw skills class")
    def student(self, name = ""):
        print("Welcome", name)
    def student(self, name = "", course = ""):
        print("Welcome", name, course)
```

```python
stud = Student()
```

```python
stud.student()
```

```python
stud.student("Billa")
```

```python
stud.student("Billa", "DS")
```

```python
# method overloading -> method in parent class and child class with some signature, the child class method will be executed

class Animal:
    def sound(self):
        print("Animal sound")
class Cat(Animal):
    def sound(self):
        print("Cat meows")
```

```python
cat = Cat()
```

```python
cat.sound()
```

ðŸ”¹ __init__ Method

It is called a constructor in Python.

It automatically runs whenever you create an object of the class.

Purpose: Initialize object variables (like setting initial values).

Example:

stud = Student("Sumit", "B.Tech")


The moment this line runs, __init__ executes.

So:

Whatever values you pass while creating the object are received by the __init__ method.

ðŸ”¹ self Keyword

Refers to the current object.

Helps store data inside the object.

Any variable using self becomes an instance attribute.

Example in your code:

self.name = name
self.degree = degree


Here:

name and degree (on the right side) are the parameters coming into __init__

self.name and self.degree store them inside the object so they can be used later

Think of self as saying:

â€œStore this value in my objectâ€

What your class does
class Student:
    def __init__(self, name, degree):
        self.name = name      # instance variable
        self.degree = degree  # instance variable


When you create objects:

s1 = Student("Rahul", "MCA")


Now,

s1.name = "Rahul"

s1.degree = "MCA"

Stored permanently in the object.

Simple Explanation Summary
Concept	Meaning
__init__	Initializes object when created
self	Refers to the object itself & stores values inside it

```python
# Private -> while the class is only accessible, use two _ to make it accessible 

class Student:
    def __init__(self, name, degree):
        self.name = name        # public attribute
        self.__degree = degree  # private attribute
        
    def show(self):
        # accessing the public and private data member
        print("name", self.name, "degree", self.__degree )
```

```python
stu = Student("Billa", "MCA")
```

```python
# Protected -> within the class and its sub class protected under it can be acessed 

class College:
    def __init__(self):
        self._college_name = "Pwskills"  # protected attribute

class Student(College):
    def __init__(self, name):
        self.name = name
        College.__init__(self)           # calling parent constructor

    def show(self):
        print("name", self.name, "college", self._college_name)
```

```python
stu = Student("Ajay")
```

```python
stu.show()
```
